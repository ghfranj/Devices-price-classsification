# Importing Libraries

```python
# Import necessary libraries for data manipulation and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Load the Prepared Data

```python
# Load the prepared data
X_train = pd.read_csv("data/X_train_prepared.csv")
y_train = pd.read_csv("data/y_train_prepared.csv")
X_test = pd.read_csv("data/X_test_prepared.csv")
```

#Summary Statistics

```python
# Summary statistics of the training data
summary_stats = X_train.describe()
summary_stats
```

#Distribution of Target Variable

```python
# Plot the distribution of the target variable
plt.figure(figsize=(8, 6))
sns.countplot(y_train)
plt.title('Distribution of Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.show()
```

#Correlation Heatmap

```python
# Calculate and plot the correlation heatmap
corr_matrix = X_train.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

#Pairplot for Key Features

```python
# Visualize relationships between key features
key_features = ['battery_power', 'ram', 'px_density', 'n_cores', 'touch_screen', 'price_range']
# Concatenate X_train and y_train to visualize relationship with price_range
train_df_concat = pd.concat([X_train, y_train], axis=1)

# Plot pairplot with hue as 'price_range'
sns.pairplot(train_df_concat[key_features], hue='price_range', diag_kind='kde')
plt.suptitle('Pairplot of Key Features')
plt.show()

```
